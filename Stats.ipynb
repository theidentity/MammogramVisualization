```python
import pandas as pd
from IPython.display import display
%matplotlib inline
```

```python
data = pd.read_csv('data/labels_converted.csv')
display(data)
```

```python
df = data.groupby('Abnormality')[['Abnormality']].agg(['count'])
display(df)
```

```python
df.plot(kind='bar')
```

```python
df = data.groupby(['Abnormality','Type'])[['Name']].count()
display(df)
```
